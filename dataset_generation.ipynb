```python
def extract_desired_sections(headings_text: str, article_text: str) -> dict:
    """
    Extracts only the sections whose headings are given in headings_text,
    for the range starting at the first heading in the list and
    ending at the last heading's section.
    
    Parameters:
      headings_text (str): Multiline string with the desired headings (one per line).
      article_text (str): The full article text that includes these headings.
      
    Returns:
      dict: A dictionary mapping each desired heading (found in the article)
            to the text block for that section.
    """
    # Create a list of desired headings (order matters)
    desired_headings = [
        line.strip() for line in headings_text.strip().splitlines() if line.strip()
    ]
    if not desired_headings:
        return {}

    first_heading = desired_headings[0]
    last_heading  = desired_headings[-1]

    sections = {}
    current_heading = None
    collecting = False
    seen_last_heading = False  # True once we have encountered the last heading

    for line in article_text.splitlines():
        stripped_line = line.strip()

        # If we haven't started collecting, look for the first desired heading.
        if not collecting:
            if stripped_line == first_heading:
                collecting = True
                current_heading = stripped_line
                sections[current_heading] = ""
                if current_heading == last_heading:
                    seen_last_heading = True
            # Otherwise ignore lines until the first desired heading appears.
            continue

        # Once collecting, check if the current line is a heading.
        if stripped_line.startswith("#"):
            if stripped_line in desired_headings:
                # It is one of the desired headings. Start a new section.
                current_heading = stripped_line
                sections[current_heading] = ""
                if stripped_line == last_heading:
                    seen_last_heading = True
            else:
                # This is a heading line that is not in our desired list.
                # If we've already seen the last desired heading, end collection.
                if seen_last_heading:
                    break
                else:
                    # Otherwise, treat it as part of the current section.
                    sections[current_heading] += line + "\n"
        else:
            # Normal content line: add to the current section.
            sections[current_heading] += line + "\n"

    # Clean up: remove any trailing whitespace from the text blocks.
    sections = {h: text.strip() for h, text in sections.items()}
    return sections



total_chars = 0
sections = extract_desired_sections(headings_text, codependent_failure)
for heading, text_block in sections.items():
    print("Heading:", heading, "length", len(text_block))
    total_chars+=len(text_block)
    print("Text Block:")
    print(text_block)
    print("-" * 40)
print(total_chars)
```
